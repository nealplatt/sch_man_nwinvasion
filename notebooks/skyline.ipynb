```python
#now extract vcfs for each population and intersect with merged baits to count the number of snps per locus
for pop in ["east_africa", "west_africa", "new_world"]:
    w_dir = "results/extended_skyline/{}".format(pop)
    if os.path.isdir(w_dir):
        shutil.rmtree(w_dir)
    os.mkdir(w_dir)

    fh  = open("{}/{}.pop".format(w_dir, pop), 'w')

    for sample in population_assignments:
        if pop is population_assignments[sample]:
            fh.write("{}\n".format(sample))
    
    fh.close()

    #now extract vcf
    vcf_cmd = "vcftools " \
        "--vcf results/variant_filtration/smv7_ex_autosomes.vcf " \
        "--keep {}/{}.pop " \
        "--mac 2 " \
        "--chr SM_V7_1 " \
        "--chr SM_V7_2 " \
        "--chr SM_V7_3 " \
        "--chr SM_V7_4 " \
        "--chr SM_V7_5 " \
        "--chr SM_V7_6 " \
        "--chr SM_V7_7 " \
        "--recode " \
        "--recode-INFO-all " \
        "--stdout " \
        ">{}/smv7_ex_autosomes_{}.vcf".format(w_dir, pop, w_dir, pop)
    
    !{vcf_cmd}

    #keep only sites genotyped in 66% of samples
    vcf_cmd = "vcftools " \
        "--vcf {}/smv7_ex_autosomes_{}.vcf " \
        "--max-missing 0.66 " \
        "--recode " \
        "--recode-INFO-all " \
        "--stdout " \
        ">{}/smv7_ex_autosomes_33m_{}.vcf".format(w_dir, pop, w_dir, pop)
    
    !{vcf_cmd}
    
        
    #remove uninformative loci from each population
    vcf_cmd = "vcftools " \
       "--vcf {}/smv7_ex_autosomes_33m_{}.vcf " \
       "--singletons " \
       "--stdout " \
       ">{}/singletons_{}.tbl".format(w_dir, pop, w_dir, pop)
    
    !{vcf_cmd}

    with open("{}/singletons_{}.tbl".format(w_dir, pop, w_dir, pop), 'r') as f:
        entries = f.readlines()
    
    with open("{}/singletons_{}.tbl".format(w_dir, pop, w_dir, pop), 'w') as f:
        for entry in entries:
            chrom, pos, sing_doub, allele, indiv = entry.split("\t")
            if (chrom != "CHROM") and (pos != "POS"):
                f.write("{}\t{}\n".format(chrom, pos))

    vcf_cmd = "vcftools " \
       "--vcf {}/smv7_ex_autosomes_{}.vcf " \
       "--exclude-positions {}/singletons_{}.tbl " \
       "--recode " \
       "--recode-INFO-all " \
       "--stdout " \
       ">{}/smv7_ex_autosomes_informative_{}.vcf".format(w_dir, pop, w_dir, pop, w_dir, pop)

    !{vcf_cmd}
    
    
    #and intersect it with exons
    bed_cmd = "bedtools intersect " \
       "-c " \
       "-a results/extended_skyline/merged_baits.bed " \
       "-b {}/smv7_ex_autosomes_informative_{}.vcf " \
       "> {}/{}_snps_per_bait.bed".format(w_dir, pop, w_dir, pop)
    
    !{bed_cmd}
    
    #now only keep loci with 6 or more informative sites
    with open("{}/{}_snps_per_bait.bed".format(w_dir, pop, w_dir, pop), 'r') as f:
        entries = f.readlines()

    with open("{}/gt_X_per_exons_{}.bed".format(w_dir, pop), 'w') as f:
        for bed_entry in entries:
            chrom, start, stop, n_snps = bed_entry.rstrip().split("\t")
            if int(n_snps) >= 6:
                f.write(bed_entry)
```

## Physically phase each locus

```python
#get locus vcf
for pop in ["east_africa", "west_africa", "new_world"]:
    w_dir = "results/extended_skyline/{}".format(pop)
    
    if os.path.isdir("{}/loci_vcfs".format(w_dir)):
        shutil.rmtree("{}/loci_vcfs".format(w_dir))
    os.mkdir("{}/loci_vcfs".format(w_dir))
    os.mkdir("{}/loci_vcfs/logs".format(w_dir))

    
    with open("{}/gt_X_per_exons_{}.bed".format(w_dir, pop), 'r') as f:
        entries = f.readlines()
    
    for entry in entries:
        chrom, start, stop, n_snps = entry.rstrip().split("\t")
        
        vcf_cmd = "vcftools " \
            "--vcf {}/smv7_ex_autosomes_informative_{}.vcf " \
            "--chr {} " \
            "--from-bp {} " \
            "--to-bp {} " \
            "--recode " \
            "--recode-INFO-all " \
            "--stdout " \
            ">{}/loci_vcfs/{}_{}_{}_{}.vcf".format(w_dir, pop, chrom, start, stop, w_dir, pop, chrom, start, stop)


        #set up qsub
        conda_cmd ="conda activate sch_man_nwinvasion-skyline"
        jid= "{}_{}_{}_{}".format(pop, chrom, start, stop)
        log= "{}/loci_vcfs/logs/{}.log".format(w_dir, jid)
        qsub_cmd = "qsub -V -cwd -S /bin/bash -q all.q -j y -N {} -o {} -pe smp 4".format(jid, log)

        #and submit to the queue
        combined_cmd = "echo \"{}; {}\" | {}".format(conda_cmd, vcf_cmd, qsub_cmd)

        #print(combined_cmd)
        !{combined_cmd}
```

```python
for pop in ["east_africa", "west_africa", "new_world"]:
    w_dir = "results/extended_skyline/{}/phased_vcfs".format(pop)
    
    if os.path.isdir(w_dir):
        shutil.rmtree(w_dir)
    os.mkdir(w_dir)
    os.mkdir("{}/logs".format(w_dir))
 
    #get bam list
    pop_file = "results/extended_skyline/{}/{}.pop".format(pop, pop)

    with open(pop_file, 'r') as f:
        entries = f.readlines()

    bam_list = []
    for entry in entries:
        bam_list.append("results/mapped_reads/{}_processed.bam".format(entry.rstrip()))

    bam_files = " ".join(bam_list)

    #glob list of vcfs
    vcfs = glob.glob("results/extended_skyline/{}/loci_vcfs/*.vcf".format(pop))

    for in_vcf in vcfs:

        abbrev = in_vcf.split("/")[-1].replace(".vcf", "")
        
        out_vcf  = "{}/{}_phased.vcf".format(w_dir, abbrev)
       
        whatshap_cmd = "whatshap phase -o {} {} {}".format(out_vcf, in_vcf, bam_files)

        #set up qsub
        conda_cmd ="conda activate sch_man_nwinvasion-whatshap"
        jid= "whats_hap_{}".format(abbrev)
        log= "{}/logs/{}.log".format(w_dir, jid)
        qsub_cmd = "qsub -V -cwd -S /bin/bash -q all.q -j y -N {} -o {} -pe smp 4".format(jid, log)

        #and submit to the queue
        combined_cmd = "echo \"{}; {}\" | {}".format(conda_cmd, whatshap_cmd, qsub_cmd)

        #print(combined_cmd)
        !{combined_cmd}
```

```python
import random

#now set up replicate runs per population
for pop in ["east_africa", "west_africa", "new_world"]:
    for rep in ["1", "2", "3"]:
        w_dir = "results/extended_skyline/{}/replicate_{}".format(pop, rep)

        if os.path.isdir(w_dir):
            shutil.rmtree(w_dir)
        os.mkdir(w_dir)
        
        #randomly select 50 loci
        fasta_files        = glob.glob("results/extended_skyline/{}/target_loci/*.fas".format(pop))
        random.shuffle(fasta_files)
        subset_fasta_files = fasta_files[0:9]
    
        #copy to rep dir
        for file in subset_fasta_files:
            dest_file = "{}/{}".format(w_dir, file.split("/")[-1])
            shutil.copyfile(file, dest_file)
            
        out_part_txt = "{}/{}_rep{}_partition.txt".format(w_dir, pop, rep)
        out_algn_nexus = "{}/{}_rep{}_alignment.nexus".format(w_dir, pop, rep)
        
        amas_cmd = "AMAS.py concat " \
            "-i {}/*.fas " \
            "-f fasta " \
            "-d dna " \
            "--out-format nexus " \
            "--concat-part {} " \
            "--concat-out {}".format(w_dir, out_part_txt, out_algn_nexus)
        
        !{amas_cmd}

        
        #clean up partition and conver to nexus
        out_part_nexus = "{}/{}_rep{}_partition.nexus".format(w_dir, pop, rep)

            
        with open(out_part_txt, 'r') as infile:
            with open(out_part_nexus, 'w') as out_file:
                out_file.write("\n#NEXUS\n\nbegin sets;\n")
                for line in infile:
                    formatted_line = "\tCHARSET {};\n".format(line.rstrip().replace(":", "-"))
                    out_file.write(formatted_line)
                out_file.write("end;\n")
        

    
        final_nexus = "{}/{}_rep{}.nexus".format(w_dir, pop, rep)
        !cat {out_algn_nexus} {out_part_nexus} >{final_nexus}
```
